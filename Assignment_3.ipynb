#4. Create an ordered list of HTML tags. Each list item must include the tag name and some information about the tag.

->  Answer:

1. <!DOCTYPE>

Defines the document type and version of HTML being used. It is placed at the very beginning of an HTML document.

Example: <!DOCTYPE html> for HTML5.

2. <html>

The root element of an HTML document. It wraps all the content on the page.

Example: <html lang="en">...</html>.

3. <head>

Contains metadata about the document, such as the title, character set, styles, and scripts. It is not displayed on the webpage.

Example: <head><title>My Page</title></head>.

4. <title>

Defines the title of the document, which appears in the browser's title bar or tab.

Example: <title>My Webpage</title>.

5. <body>

Contains the visible content of the webpage, such as text, images, links, and other elements.

Example: <body><h1>Hello World</h1></body>.

6. <h1> to <h6>

Headings for the document, with <h1> being the highest level and <h6> the lowest.

Example: <h1>Main Heading</h1>.

7. <p>

Defines a paragraph of text.

Example: <p>This is a paragraph.</p>.

8. <a>

Creates a hyperlink to another webpage, file, or location within the same page.

Example: <a href="https://example.com">Visit Example</a>.

9. <img>

Embeds an image in the document. It is a self-closing tag.

Example: <img src="image.jpg" alt="Description">.

10. <ul>

Defines an unordered (bulleted) list.

Example: <ul><li>Item 1</li><li>Item 2</li></ul>.

11. <ol>

Defines an ordered (numbered) list.

Example: <ol><li>First</li><li>Second</li></ol>.

12. <li>

Represents a list item within <ul> or <ol>.

Example: <li>List Item</li>.

13. <div>

A block-level container used to group elements for styling or layout purposes.

Example: <div>Content here</div>.

14. <span>

An inline container used to style or group small portions of text.

Example: <span style="color:red;">Red Text</span>.

15. <table>

Creates a table to display data in rows and columns.

Example: <table><tr><th>Header</th></tr><tr><td>Data</td></tr></table>.

16. <tr>

Defines a row in a table.

Example: <tr><td>Row Data</td></tr>.

17. <td>

Defines a cell in a table.

Example: <td>Cell Data</td>.

18. <th>

Defines a header cell in a table.

Example: <th>Header</th>.

19. <form>

Creates a form for user input, such as text fields, checkboxes, and buttons.

Example: <form action="/submit" method="post"><input type="text"></form>.

20. <input>

Creates an input field for user interaction. It can be of various types like text, password, checkbox, etc.

Example: <input type="text" placeholder="Enter text">.

21. <button>

Creates a clickable button.

Example: <button>Click Me</button>.

22. <label>

Associates a label with a form element, improving accessibility.

Example: <label for="name">Name:</label><input id="name" type="text">.

23. <textarea>

Creates a multi-line text input field.

Example: <textarea rows="4" cols="50">Enter text here</textarea>.

24. <select>

Creates a dropdown list for user selection.

Example: <select><option>Option 1</option><option>Option 2</option></select>.

25. <option>

Represents an option in a dropdown list (<select>).

Example: <option value="1">Option 1</option>.

26. <header>

Represents the header section of a document or section.

Example: <header><h1>Welcome</h1></header>.

27. <footer>

Represents the footer section of a document or section.

Example: <footer><p>Copyright 2023</p></footer>.

28. <nav>

Defines a section of navigation links.

Example: <nav><a href="/">Home</a></nav>.

29. <section>

Represents a standalone section of content.

Example: <section><h2>About</h2><p>Details here</p></section>.

30. <article>

Represents a self-contained piece of content, such as a blog post or news article.

Example: <article><h2>Blog Title</h2><p>Content</p></article>.

31. <aside>

Represents content that is tangentially related to the main content, such as a sidebar.

Example: <aside><p>Related Info</p></aside>.

32. <meta>

Provides metadata about the HTML document, such as character encoding or viewport settings.

Example: <meta charset="UTF-8">.
